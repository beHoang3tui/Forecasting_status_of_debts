- Data

```python
X = []
for i in range(10000):
  X.append([df['Current Loan Amount'][i] , df['Monthly Debt'][i] , df['Years of Credit History'][i] , df['Number of Open Accounts'][i] , df['Current Credit Balance'][i]])
Y = np.array([1 if i =='Fully Paid' else 0 for i in df['Loan Status'][:10000]])
```

- Split data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import  classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression 
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def split_data(ratio : float):
  x_train,x_test,y_train,y_test = train_test_split(X, Y ,test_size= ratio,random_state=42)
  return x_train,x_test,y_train,y_test
```

### **- Chia tập dữ liệu theo các tỷ lệ giữa tập train và tập test : 80:20**

```python
x_train_1,x_test_1,y_train_1 , y_test_1 = split_data(0.2)
```

**Logistic Regression**

```python
classifier_1= LogisticRegression()
classifier_1.fit(x_train_1, y_train_1)
```

```python
len(x_train_1)
```

Predict

```python
y_pred_1 = classifier_1.predict(x_test_1)
print('Classifcation report:\n', classification_report(y_test_1, y_pred_1))
conf_mat_1 = confusion_matrix(y_true=y_test_1, y_pred=y_pred_1)
print('Confusion matrix:\n', conf_mat_1)
```

**KNN**

```python
neigh_1 = KNeighborsClassifier(n_neighbors=3)
neigh_1.fit(x_train_1, y_train_1)
```

Predict

```python
y_pred_2 = neigh_1.predict(x_test_1)
print('Classifcation report:\n', classification_report(y_test_1, y_pred_2))
conf_mat_2 = confusion_matrix(y_true=y_test_1, y_pred=y_pred_2)
print('Confusion matrix:\n', conf_mat_2)
```

**Decision tree**

```python
clf_1 = tree.DecisionTreeClassifier()
clf_1.fit(x_train_1, y_train_1)
```

Predict

```python
y_pred_3 = clf_1.predict(x_test_1)
print('Classifcation report:\n', classification_report(y_test_1, y_pred_3))
conf_mat_3 = confusion_matrix(y_true=y_test_1, y_pred=y_pred_3)
print('Confusion matrix:\n', conf_mat_3)
```

**Random forest**

```python
clf_2 = RandomForestClassifier(n_estimators=20,criterion='entropy', random_state=0,max_depth=10)
clf_2.fit(x_train_1, y_train_1)
```

Predict
